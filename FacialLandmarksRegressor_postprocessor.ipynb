```python
import cv2
import degirum as dg
from custom_postprocessors.faciallandmarksregressor import FacialLandmarkRegressor
```

```python
zoo = dg.connect_model_zoo("0.0.0.0:8778")
model = zoo.load_model("landmarks_regression_retail--62x62_float_openvino_cpu_1")
model.custom_postprocessor = FacialLandmarkRegressor
```

```python
image_path = "face_crop.jpg" 
image = cv2.imread(image_path) # read the cropped face image
faceheight, facewidth, channels = image.shape 
res= model(image)
output=res.results
output
```

```python
for det in output[0]["landmarks"]:
    cv2.circle(image, (det["landmark"][0],det["landmark"][1]),2,(0,0,255),-1)    
cv2.imwrite(image_path.split(".")[0]+'_output.jpg', image)    
```
