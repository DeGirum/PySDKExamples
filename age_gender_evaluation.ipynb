```python
import json
import os
import degirum as dg
zoo = dg.connect_model_zoo("0.0.0.0:8778")
model = zoo.load_model("age_gender_recognition_retail--62x62_float_openvino_cpu_1")
model.measure_time=True
```

* `DATASET PATH` : Path to the face dataset with annotations of Age and Gender. This model will recognize age of people in [18, 75] years old range and it is not applicable for children.



* `INPUT_DATA.json` : It contains a list of dictionaries that contain image file name, gender category ( 0 - Male, 1 - Female) and age (integer or float).

For example:

    [
    
    {'file_name': '19_0_0_20170102233014401.jpg.chip.jpg','gender': '0','age': '19'},
    
    {'file_name': '19_0_0_20170103180141544.jpg.chip.jpg','gender': '0','age': '19'},
    
    ...
    
    ]

```python
input_file_path = '../age_gender/age_gender_input.json'
dataset_path = "../age_gender/UTKFace/"
```
