```python
import degirum as dg
from google.colab.patches import cv2_imshow

zoo = dg.connect(dg.CLOUD, zoo_url="https://cs.degirum.com/degirum/ultralytics", token="<insert token here>")
model = zoo.load_model("yolov8n_silu_coco--640x640_quant_tflite_edgetpu_1")
model.image_backend = "opencv"
res = model("https://raw.githubusercontent.com/DeGirum/PySDKExamples/main/images/TwoCats.jpg")
cv2_imshow(res.image_overlay)
```
