 The below script has a few parameters :
 
 1. `OutputConfThreshold`  - Output Confidence threshold (default : 0.001)

 2. `OutputNMSThreshold` - Output Non-Max Suppression thershold (default : 0.6)

 3. `MaxDetections` - Maximum Detections (default : 100)
 
 4. `MaxDetectionsPerClass` - Maximum Detections Per Class (default : 20)
 
 5. `MaxClassesPerDetection` - Maximum Classes Per Detection (default : 1)
 
 6. `UseRegularNMS` - Whether to use Regular Non-Max Suppression (default : True)
 
 7. `InputResizeMethod` - Input Resize Method (default : Bilinear)
 
 8. `InputPadMethod` - Input Pad Method (default : letterbox)
 
 9. `ImageBackend` - Image Backend (default : opencv)
 
 10. `InputImgFmt` - InputImgFmt (default : JPEG)
